## Visualization


```python
#TV vs Sales
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(sales_data['TV'], sales_data['Sales'], color='blue')
plt.title("TV vs Sales")
plt.xlabel("TV")
plt.ylabel("Sales")

#Radio vs Sales
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 2)
plt.scatter(sales_data['Radio'],sales_data["Sales"],c='blue')
plt.title("Radio vs Sales")
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.show()

#Newspaper vs Sales
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 3)
plt.scatter(sales_data['Newspaper'],sales_data["Sales"],c='blue')
plt.title("Newspaper vs Sales")
plt.xlabel('Newspaper')
plt.ylabel('Sales')
plt.show()


plt.tight_layout()
plt.show()
```

```python
# Histogram of Sales
plt.hist(sales_data['Sales'], color="green", edgecolor='white', bins=10)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
#Pairplot
sns.set(style="darkgrid")
sns.pairplot(sales_data)
plt.show()
```

Sales Highly coorelated with the tv

train our model using linear Regression as it is coorelated with only one variable TV

```python
x_train,x_test,y_train,y_test=train_test_split(sales_data[['TV']],sales_data[['Sales']],test_size=0.3,random_state=0)
```

```python
#Train Regresion model
model1=LinearRegression()
model1.fit(x_train,y_train)
```

```python
#Predicction
result=model1.predict(x_test)
print(result)
```

```python
#Model Coefficient
print("Model Coefficent:-",model1.coef_)
print("Model Intercept:-",model1.intercept_)
```

```python
# Model Evaluation
print("Mean Squared Error:", mean_squared_error(y_test, result))
print("R-squared Score:", r2_score(y_test, result))
```

```python
# Plot Predictions
plt.scatter(x_test, y_test, label="Actual Data", color='blue')
plt.plot(x_test, result, color='red', label="Regression Line")
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("TV vs Sales (Regression Model)")
plt.legend()
plt.show()
```

```python
X = sales_data[['TV', 'Radio', 'Newspaper']]
y = sales_data[['Sales']]
```

```python
pred_multi = model2.predict(x_test)
```

```python
print("Multiple Linear Regression Coefficients:", model2.coef_)
print("Multiple Linear Regression Intercept:", model2.intercept_)
print("Multiple Regression Mean Squared Error:", mean_squared_error(y_test, pred_multi))
print("Multiple Regression R-squared Score:", r2_score(y_test, pred_multi))
```
